```python
#Imports
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.ticker as mtick
import matplotlib.style as style
from matplotlib import cm
```

2. Next run the PCI Plot Function cell below

```python
#PCI Plot function 
def PCIplot(df, Title, ylabel, xlabel):
  x = df.variable.to_list()
  y = df.Mean.to_list()
  size = df.PCI.to_list()
  mutipliedsize = [element * 1000 for element in size]
  
  fig, ax = plt.subplots(figsize = (7,5),dpi = 200) #Figure size
  plt.style.use('seaborn-bright')
  Titlefont = {'family': 'Times New Roman', 
            'color':  'black',
            'weight': 'normal',
            'size': 16,}
  #set fomt size for axes
  AxesFont= {'family': 'Times New Roman',
            'color':  'black',
            'weight': 'normal',
            'size': 14,}
  ax.spines['right'].set_visible(False)
  ax.spines['top'].set_visible(False)
  plt.plot(x,y, zorder = 1, alpha = 0.7,c = '0.4', ls=":")
  ramp = df.PCI.to_list()
  plt.scatter(x,y,s = mutipliedsize,zorder =2,c = ramp, cmap = 'coolwarm')
  for i in range(len(size)):
    if i <=1:
      ax.annotate(str('Mean: '+ str(df.loc[i][2])+ "\n PCI: "+ str(df.loc[i][1])[1:]), xy= (df.loc[i][0], df.loc[i][2]), xytext = (10,-45),
                  bbox=dict(boxstyle = 'round', fc='0.9',lw ='0' ), arrowprops= dict(arrowstyle = '-',color="0.6", ls = ":", shrinkA=0, shrinkB=0,
                      connectionstyle="angle,angleA=0,angleB=90,rad=10"),textcoords = 'offset points', size = 7)
    else:
      ax.annotate(str('Mean: '+ str(df.loc[i][2])+ "\n PCI: "+ str(df.loc[i][1])[1:]), xy= (df.loc[i][0], df.loc[i][2]), xytext = (-50,30),
              bbox=dict(boxstyle = 'round', fc='0.9',lw ='0' ), arrowprops= dict(arrowstyle = '-',color="0.6", ls = ":", shrinkA=0, shrinkB=0,
                  connectionstyle="angle,angleA=0,angleB=90,rad=10"),textcoords = 'offset points', size = 7)
  
  
  xlabellist=[]
  for i in range(len(df.variable)):
      xlabellist.append(df.variable[i])
  plt.xticks(df.variable)
  plt.ylim([-4,4.5])
  plt.yticks(np.arange(-4,4.5,1),np.arange(-4,4.2,1))
  plt.axhline(y = 0.0, color = 'r', linestyle = '-', alpha = .3)
  plt.title(Title, fontdict = Titlefont)
  plt.ylabel(ylabel, fontdict = AxesFont)
  plt.xlabel(xlabel, fontdict = AxesFont)
  #plt.colorbar()
  plt.tight_layout()
  plt.show()
```

```python
def MultiplePCIplot(df, Title, ylabel, xlabel):
  fig, ax = plt.subplots(figsize = (8,5),dpi = 200) #Figure size
  ax.spines['right'].set_visible(False)
  ax.spines['top'].set_visible(False)
  plt.style.use('seaborn-bright')
  Titlefont = {'family': 'Times New Roman', 
            'color':  'black',
            'weight': 'normal',
            'size': 16,}
  #set fomt size for axes
  AxesFont= {'family': 'Times New Roman',
            'color':  'black',
            'weight': 'normal',
            'size': 14,}
  sns.lineplot(
      x = df.variable,
      y = df.Mean,
      data = df,
      hue = df.RaceEthnicity,
      ci = False,
      legend = 'auto',
      palette = 'viridis'
  )
  sns.scatterplot(
      x= df.variable,
      y = df.Mean,
      data = df,
      size = df.PCI,
      sizes = (30,350),
      hue = df.PCI,
      legend = 'auto',
      palette = 'coolwarm'

   )
  xlabellist=[]
  for i in range(len(df.variable)):
      xlabellist.append(df.variable[i])
  plt.xticks(np.arange(min(xlabellist),max(xlabellist)+1,1))
 
  plt.ylim([-4,4.5])
  plt.yticks(np.arange(-4,4.5,1),np.arange(-4,4.2,1))
  plt.axhline(y = 0.0, color = 'r', linestyle = '-', alpha = .3)
  plt.title(Title, fontdict = Titlefont)
  plt.ylabel(ylabel, fontdict = AxesFont)
  plt.xlabel(xlabel, fontdict = AxesFont)
  plt.legend(title = 'Legend:', title_fontsize ='x-large', fontsize = 'large',
             bbox_to_anchor=(1.02, 1),loc='upper left', borderaxespad=0)
  plt.tight_layout()
  plt.show()
```

```python
MultiplePCIplot(PAOT_df, 'PAOT by Race/Ethnicity','Acceptability','PAOT')
```

```python
MultiplePCIplot(BAOT_df, 'BAOT by Race/Ethnicity','Acceptability','BAOT')
```
